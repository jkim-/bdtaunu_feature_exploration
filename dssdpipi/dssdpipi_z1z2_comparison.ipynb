## Comparison of PDF shapes between "gap" D** signal MC samples

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from IPython.display import HTML
```

```python
d1 = np.genfromtxt('d1.csv', skip_header=1, delimiter=',')
d2s = np.genfromtxt('d2s.csv', skip_header=1, delimiter=',')
```

```python
plt.figure(figsize=(10,6))
ax = plt.gca()
ax.hist([d1[:,0], d2s[:,0]], bins='auto', normed=True, histtype='step', alpha=0.8,
        label = [r'$B\rightarrow{}D_1(D\pi\pi)l\nu$', r'$B\rightarrow{}D(2S)^{(*)}(D\pi\pi)l\nu$']);
ax.set_xlim([-3.5,0]);
ax.set_title(r'PDFs of $z_1$ for $D_1$ and $D(2S)^{(*)}$ samples', fontsize=24);
ax.set_xlabel(r'$z_1$', fontsize=20);
ax.legend(prop={'size':15});
ax.tick_params(length=8, width=1, labelsize=16);
ttl = ax.title;
ttl.set_position([.5, 1.05]);
```

```python
plt.figure(figsize=(10,6))
ax = plt.gca()
ax.hist([d1[:,1], d2s[:,1]], bins='auto', normed=True, histtype='step', alpha=0.8,
        label = [r'$B\rightarrow{}D_1(D\pi\pi)l\nu$', r'$B\rightarrow{}D(2S)^{(*)}(D\pi\pi)l\nu$']);
ax.set_xlim([-1.5,2]);
ax.set_title(r'PDFs of $z_2$ for $D_1$ and $D(2S)^{(*)}$ samples', fontsize=24);
ax.set_xlabel(r'$z_2$', fontsize=20);
ax.legend(prop={'size':15});
ax.tick_params(length=8, width=1, labelsize=16);
ttl = ax.title;
ttl.set_position([.5, 1.05]);
```
